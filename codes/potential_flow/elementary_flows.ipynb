```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
u = np.cos(np.pi/4) + 0*X
v = np.sin(np.pi/4) + 0*X

fig, ax = plt.subplots()

# draw streamplot with explicit low zorder
sp = ax.streamplot(X, Y, u, v, density=0.5)
ax.plot([-1, 1], [0, 0], color='lightgray', linewidth=2, zorder=0)
ax.plot([0, 0], [-1, 1], color='lightgray', linewidth=2, zorder=0)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12, rotation=0, labelpad=15)
ax.set_title('Uniform flow')

plt.show()
fig.savefig('uniformflow.pdf')
```

```python
u = X
v = -Y

fig, ax = plt.subplots()

# draw streamplot with explicit low zorder
sp = ax.streamplot(X, Y, u, v, density=0.8)
ax.plot([-1, 1], [0, 0], color='lightgray', linewidth=2, zorder=0)
ax.plot([0, 0], [-1, 1], color='lightgray', linewidth=2, zorder=0)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12, rotation=0, labelpad=15)
ax.set_title('Stagnation point flow')

plt.show()
fig.savefig('stagnationflow.pdf')
```

```python
rsq = X**2 + Y**2;
u = 1./(2*np.pi*rsq)*X
v = 1./(2*np.pi*rsq)*Y

fig, ax = plt.subplots()

# draw streamplot with explicit low zorder
sp = ax.streamplot(X, Y, u, v, density=0.8)
ax.plot([-1, 1], [0, 0], color='lightgray', linewidth=2, zorder=0)
ax.plot([0, 0], [-1, 1], color='lightgray', linewidth=2, zorder=0)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12, rotation=0, labelpad=15)
ax.set_title('Line source')

plt.show()
fig.savefig('linesource.pdf')
```

```python
rsq = X**2 + Y**2;
u = 1 + 1./(2*np.pi*rsq)*X
v = 1./(2*np.pi*rsq)*Y

fig, ax = plt.subplots()

# draw streamplot with explicit low zorder
sp = ax.streamplot(X, Y, u, v, density=1)
ax.plot([-1, 1], [0, 0], color='lightgray', linewidth=2, zorder=0)
ax.plot([0, 0], [-1, 1], color='lightgray', linewidth=2, zorder=0)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12, rotation=0, labelpad=15)
ax.set_title('Line source')

plt.show()
fig.savefig('movinglinesource.pdf')
```

```python
# Radial distance
r2 = X**2 + Y**2
r2[r2 == 0] = 1e-10  # avoid divide by zero

# Velocity field (tangential vortex flow)
U = -Y / (2 * np.pi * r2)
V =  X / (2 * np.pi * r2)

# Plot streamlines with arrows
plt.figure(figsize=(6,6))
plt.streamplot(X, Y, U, V, density=1, arrowsize=1.2, linewidth=1, color="navy")

# Mark the singularity at the origin
plt.scatter(0, 0, color="red", s=50, label="Origin (singularity)")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Streamlines of $\\psi = -\\frac{1}{2\\pi} \\ln r$")
plt.legend()
plt.axis("equal")
plt.show()
```
