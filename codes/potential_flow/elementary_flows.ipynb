```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
u = np.cos(np.pi/4) + 0*X
v = np.sin(np.pi/4) + 0*X

fig, ax = plt.subplots()

# draw streamplot with explicit low zorder
sp = ax.streamplot(X, Y, u, v, density=0.5)
ax.plot([-1, 1], [0, 0], color='lightgray', linewidth=2, zorder=0)
ax.plot([0, 0], [-1, 1], color='lightgray', linewidth=2, zorder=0)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12, rotation=0, labelpad=15)
ax.set_title('Uniform flow')

plt.show()
fig.savefig('uniformflow.pdf')
```

```python
u = X
v = -Y

fig, ax = plt.subplots()

# draw streamplot with explicit low zorder
sp = ax.streamplot(X, Y, u, v, density=0.8)
ax.plot([-1, 1], [0, 0], color='lightgray', linewidth=2, zorder=0)
ax.plot([0, 0], [-1, 1], color='lightgray', linewidth=2, zorder=0)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12, rotation=0, labelpad=15)
ax.set_title('Stagnation point flow')

plt.show()
fig.savefig('stagnationflow.pdf')
```

```python
rsq = X**2 + Y**2;
u = 1./(2*np.pi*rsq)*X
v = 1./(2*np.pi*rsq)*Y

fig, ax = plt.subplots()

# draw streamplot with explicit low zorder
sp = ax.streamplot(X, Y, u, v, density=0.8)
ax.plot([-1, 1], [0, 0], color='lightgray', linewidth=2, zorder=0)
ax.plot([0, 0], [-1, 1], color='lightgray', linewidth=2, zorder=0)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12, rotation=0, labelpad=15)
ax.set_title('Line source')

plt.show()
fig.savefig('linesource.pdf')
```
